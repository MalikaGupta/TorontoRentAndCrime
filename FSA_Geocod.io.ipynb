```python
import pandas as pd
import requests
import time

# Load Crime Statistics data
crime_statistics_file_path = r"C:/Users/malik/OneDrive/Desktop/BI/CaseStudy/Dataset/Crime_Statistics_Current.xlsx"
crime_statistics_df = pd.read_excel(crime_statistics_file_path,sheet_name='Sheet2')

# Initialize the Geocod.io API with your API key
api_key = 'a762a88227667b712576b671685b2165772a775'  # Replace this with your actual Geocod.io API key

# Function to get the postal code from latitude and longitude using Geocod.io API
def get_postal_code(lat, lon):
    try:
        url = f"https://api.geocod.io/v1.7/reverse?q={lat},{lon}&api_key={api_key}"
        response = requests.get(url)
        data = response.json()
        #print(f"API Response for {lat}, {lon}: {data}")  # Debugging line
        if 'results' in data and len(data['results']) > 0:
            address_components = data['results'][0]['address_components']
            return address_components['zip']
        return None
    except Exception as e:
        print(f"Error retrieving postal code for {lat}, {lon}: {e}")
        return None

# Apply the function to the Crime Statistics dataset with progress monitoring
total_rows = len(crime_statistics_df)
for index, row in crime_statistics_df.iterrows():
    lat, lon = row['Latitude'], row['Longitude']
    fsa = get_postal_code(lat, lon)
    crime_statistics_df.at[index, 'FSA'] = fsa
    
    # Print progress
    if index % 100 == 0 or index == total_rows - 1:
        print(f"Processed {index + 1}/{total_rows} rows")
    
    # Respect the rate limit
    time.sleep(0.1)  # Adjust the sleep time if needed

# Display the first few rows to verify the data
print("Crime Statistics Data with FSA:")
print(crime_statistics_df.head())


# Save the updated Crime Statistics dataset with FSA included
output_file_path = r'C:/Users/malik/OneDrive/Desktop/BI/CaseStudy/Dataset/FSA_Crime.csv'
crime_statistics_df.to_csv(output_file_path, index=False)

print("Updated Crime Statistics data saved successfully with FSA.")
```
